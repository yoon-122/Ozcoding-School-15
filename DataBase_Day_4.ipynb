## ë¬¸ì œ 2: í…Œì´ë¸” ìƒì„± (CREATE TABLE)

ë‹¤ìŒ 5ê°œì˜ í…Œì´ë¸”ì„ ìƒì„±í•©ë‹ˆë‹¤:
1. PetOwners - ë°˜ë ¤ë™ë¬¼ ì£¼ì¸ í…Œì´ë¸”
2. Pets - ë°˜ë ¤ë™ë¬¼ í…Œì´ë¸”
3. Rooms - ê°ì‹¤ í…Œì´ë¸”
4. Reservations - ì˜ˆì•½ í…Œì´ë¸”
5. Services - ì„œë¹„ìŠ¤ í…Œì´ë¸”

```python
# â‘  ë°˜ë ¤ë™ë¬¼ ì£¼ì¸ í…Œì´ë¸” (PetOwners Table)
# ë¬¸ì œ: ë°˜ë ¤ë™ë¬¼ í˜¸í…” ì˜ˆì•½ ì‹œìŠ¤í…œì—ì„œ ê³ ê°(ë°˜ë ¤ë™ë¬¼ ì£¼ì¸) ì •ë³´ë¥¼ ì €ì¥í•˜ëŠ” í…Œì´ë¸” ìƒì„±

create_pet_owners = """
CREATE TABLE PetOwners (
    ownerID INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    contact TEXT
);
"""

cursor.execute(create_pet_owners)
print("âœ… PetOwners í…Œì´ë¸” ìƒì„± ì™„ë£Œ!")
print("   - ownerID: ì£¼ì¸ ID (ê¸°ë³¸ í‚¤, ìë™ ì¦ê°€)")
print("   - name: ì£¼ì¸ ì´ë¦„ (í•„ìˆ˜)")
print("   - contact: ì—°ë½ì²˜")
```

```python
# â‘¡ ë°˜ë ¤ë™ë¬¼ í…Œì´ë¸” (Pets Table)
# ë¬¸ì œ: ë°˜ë ¤ë™ë¬¼ ì •ë³´ë¥¼ ì €ì¥í•˜ëŠ” í…Œì´ë¸” ìƒì„±
# í•œ ê³ ê°ì´ ì—¬ëŸ¬ ë°˜ë ¤ë™ë¬¼ì„ ê°€ì§ˆ ìˆ˜ ìˆìœ¼ë¯€ë¡œ PetOwnersì™€ 1:N ê´€ê³„

create_pets = """
CREATE TABLE Pets (
    petID INTEGER PRIMARY KEY AUTOINCREMENT,
    ownerID INTEGER NOT NULL,
    name TEXT NOT NULL,
    species TEXT,
    breed TEXT,
    FOREIGN KEY (ownerID) REFERENCES PetOwners(ownerID)
);
"""

cursor.execute(create_pets)
print("âœ… Pets í…Œì´ë¸” ìƒì„± ì™„ë£Œ!")
print("   - petID: ë°˜ë ¤ë™ë¬¼ ID (ê¸°ë³¸ í‚¤, ìë™ ì¦ê°€)")
print("   - ownerID: ì£¼ì¸ ID (ì™¸ë˜ í‚¤ â†’ PetOwners)")
print("   - name: ë°˜ë ¤ë™ë¬¼ ì´ë¦„ (í•„ìˆ˜)")
print("   - species: ì¢…ë¥˜ (ê°•ì•„ì§€, ê³ ì–‘ì´ ë“±)")
print("   - breed: í’ˆì¢…")
```

```python
# â‘¢ ê°ì‹¤ í…Œì´ë¸” (Rooms Table)
# ë¬¸ì œ: í˜¸í…” ê°ì‹¤ ì •ë³´ë¥¼ ì €ì¥í•˜ëŠ” í…Œì´ë¸” ìƒì„±

create_rooms = """
CREATE TABLE Rooms (
    roomID INTEGER PRIMARY KEY AUTOINCREMENT,
    roomNumber TEXT NOT NULL UNIQUE,
    roomType TEXT,
    pricePerNight REAL NOT NULL
);
"""

cursor.execute(create_rooms)
print("âœ… Rooms í…Œì´ë¸” ìƒì„± ì™„ë£Œ!")
print("   - roomID: ê°ì‹¤ ID (ê¸°ë³¸ í‚¤, ìë™ ì¦ê°€)")
print("   - roomNumber: ê°ì‹¤ ë²ˆí˜¸ (í•„ìˆ˜, ê³ ìœ ê°’)")
print("   - roomType: ê°ì‹¤ íƒ€ì… (ìŠ¤íƒ ë‹¤ë“œ, ë””ëŸ­ìŠ¤, ìŠ¤ìœ„íŠ¸ ë“±)")
print("   - pricePerNight: 1ë°• ê°€ê²© (í•„ìˆ˜)")
```

```python
# â‘£ ì˜ˆì•½ í…Œì´ë¸” (Reservations Table)
# ë¬¸ì œ: ë°˜ë ¤ë™ë¬¼ í˜¸í…” ì˜ˆì•½ ì •ë³´ë¥¼ ì €ì¥í•˜ëŠ” í…Œì´ë¸” ìƒì„±
# í•œ ë°˜ë ¤ë™ë¬¼ì€ ì—¬ëŸ¬ ë²ˆ ì˜ˆì•½í•  ìˆ˜ ìˆê³ , ì˜ˆì•½ì€ ê°ì‹¤ê³¼ ì—°ê²°ë˜ì–´ì•¼ í•¨

create_reservations = """
CREATE TABLE Reservations (
    reservationID INTEGER PRIMARY KEY AUTOINCREMENT,
    petID INTEGER NOT NULL,
    roomID INTEGER NOT NULL,
    checkInDate TEXT NOT NULL,
    checkOutDate TEXT NOT NULL,
    FOREIGN KEY (petID) REFERENCES Pets(petID),
    FOREIGN KEY (roomID) REFERENCES Rooms(roomID)
);
"""

cursor.execute(create_reservations)
print("âœ… Reservations í…Œì´ë¸” ìƒì„± ì™„ë£Œ!")
print("   - reservationID: ì˜ˆì•½ ID (ê¸°ë³¸ í‚¤, ìë™ ì¦ê°€)")
print("   - petID: ë°˜ë ¤ë™ë¬¼ ID (ì™¸ë˜ í‚¤ â†’ Pets)")
print("   - roomID: ê°ì‹¤ ID (ì™¸ë˜ í‚¤ â†’ Rooms)")
print("   - checkInDate: ì²´í¬ì¸ ë‚ ì§œ (í•„ìˆ˜)")
print("   - checkOutDate: ì²´í¬ì•„ì›ƒ ë‚ ì§œ (í•„ìˆ˜)")
```

```python
# â‘¤ ì„œë¹„ìŠ¤ í…Œì´ë¸” (Services Table)
# ë¬¸ì œ: ì˜ˆì•½ ì‹œ í•¨ê»˜ ì˜ˆì•½í•  ìˆ˜ ìˆëŠ” ì¶”ê°€ ì„œë¹„ìŠ¤ ì •ë³´ë¥¼ ì €ì¥í•˜ëŠ” í…Œì´ë¸” ìƒì„±
# ì˜ˆì•½ ì‹œ ë‹¤ì–‘í•œ ì„œë¹„ìŠ¤ë„ í•¨ê»˜ ì˜ˆì•½í•  ìˆ˜ ìˆìœ¼ë¯€ë¡œ Reservationsì™€ 1:N ê´€ê³„

create_services = """
CREATE TABLE Services (
    serviceID INTEGER PRIMARY KEY AUTOINCREMENT,
    reservationID INTEGER NOT NULL,
    serviceType TEXT NOT NULL,
    price REAL NOT NULL,
    FOREIGN KEY (reservationID) REFERENCES Reservations(reservationID)
);
"""

cursor.execute(create_services)
print("âœ… Services í…Œì´ë¸” ìƒì„± ì™„ë£Œ!")
print("   - serviceID: ì„œë¹„ìŠ¤ ID (ê¸°ë³¸ í‚¤, ìë™ ì¦ê°€)")
print("   - reservationID: ì˜ˆì•½ ID (ì™¸ë˜ í‚¤ â†’ Reservations)")
print("   - serviceType: ì„œë¹„ìŠ¤ ì¢…ë¥˜ (í•„ìˆ˜)")
print("   - price: ê°€ê²© (í•„ìˆ˜)")
```

```python
# ìƒì„±ëœ í…Œì´ë¸” ëª©ë¡ í™•ì¸
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()

print("ğŸ“‹ ìƒì„±ëœ í…Œì´ë¸” ëª©ë¡:")
for table in tables:
    print(f"   - {table[0]}")

print(f"\nâœ… ì´ {len(tables)}ê°œì˜ í…Œì´ë¸”ì´ ìƒì„±ë˜ì—ˆìŠµë‹ˆë‹¤!")
```

```python
# ëª¨ë“  ì˜ˆì•½ ì •ë³´ ì¡°íšŒ (ë°˜ë ¤ë™ë¬¼, ì£¼ì¸, ê°ì‹¤ ì •ë³´ í¬í•¨)
query_reservations = """
SELECT
    r.reservationID,
    po.name AS owner_name,
    p.name AS pet_name,
    p.species,
    rm.roomNumber,
    rm.roomType,
    r.checkInDate,
    r.checkOutDate,
    rm.pricePerNight
FROM Reservations r
JOIN Pets p ON r.petID = p.petID
JOIN PetOwners po ON p.ownerID = po.ownerID
JOIN Rooms rm ON r.roomID = rm.roomID;
"""

df_result = pd.read_sql_query(query_reservations, conn)
print("ğŸ“Š ëª¨ë“  ì˜ˆì•½ ì •ë³´:")
print(df_result.to_string(index=False))
```
